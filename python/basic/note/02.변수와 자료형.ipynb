### 문자열 포맷팅
: %d(decimal), %s(string), %f(float) => (전통적인 방법)

```python
# %d : 정수형 format
appleCount = 2
print("I have %d apples."%appleCount)
print("I have", appleCount,"apples.")
# %s : 문자열
appleCount = "two"
print("I have %s apples."%appleCount)

# 혼합형 (중가로 사용)
number = 2
fruit = "apples"
print("I have %d %s."%(number,fruit))


```

```python
# 문자열 정렬 : 10글자가 들어갈거다
print("-"*30)
print("I have 2 %10s."%"apples")  # 총글자 10자리 우측정렬
print("I have 2 %-10s."%"apples") # 총글자 10자리 좌측정렬
```

```python
# %f 실수형 format 
print("-"*30)
print("%.4f"%3.14159)   # 실수형 소숫점 4자리
print("%10.4f"%3.14159) # 실수형 소숫점 포함 총 10자리


```

### 문자열 관련 함수

```python
# package pylance가 해줌

str = "I have a dream."
#글자 갯수 (한글도 한글자수)
print(len(str))

# string의 내부 함수 count 
print(str.count('a'))
print(str.count(' '))

# FIND: 문자 위치 (index)
print(str.find('e')) # 5번째 e가 위치해 있음.
print(str.find('x')) # 없으면 -1
print(str[str.find('e')]) # 인덱스를 이용해서 value를 확인

# e가 두개 있음. pylance
print(str.find('e',6)) 

# 특정 글자의 모든 위치를 알아내는 방법 -1
str = "I have a dream."
search_word = 'a'
start = 0
count = 0

# 무한반복: 단어찾기
while True:
  position = str.find(search_word,start)
  if position == -1:
    break
  count+=1
  print(f"단어 {search_word}의 {count}번 위치는 {position} 입니다.")
  start = position + len(search_word)



```

```python

# 특정 글자의 모든 위치를 알아내는 방법 - 2 : 정규식
import re

str = "I have a dream."
search_word = 'a'
for match in re.finditer(search_word, str) :
  print(match.start(),match.end())

# 컨프리헨션 comprehension
[match.start() for match in re.finditer(search_word, str)]
```

### 날짜형 변환

```python
from datetime import datetime, timedelta

time1 = datetime(2025,12,25,21,0,0)
print(time1)
print(type(time1))

# 날짜 계산하기
time2 = datetime.now()
print(time2)

print(f"일: {(time2-time1).days}, 초: {(time2-time1).total_seconds()//(24*60*60)}")
print((f"{time2-time1}").split(",")[0])
```

```python
# 특정 날짜로 부터 시간 계산
timeOn = datetime.now()

print(timeOn)
print('현재 시간부터 5일 뒤')
print(timeOn+timedelta(days=5))
print('현재 시간부터 3일 전')
print(timeOn+timedelta(days=-3))
print('현재 시간부터 1일뒤의 2시간 전')
print(timeOn+timedelta(days=1,hours=-2))
```

```python
# 현재 일자
print(timeOn.date())
print(timeOn.date().year)

# 올해 연말까지 몇일이 남았을까?
print(f"{(datetime(timeOn.date().year+1,1,1,0,0,0)-timeOn).days}일 남았습니다")




```

### 리스트 관련 함수들

```python
# 정수 추가 하기
var = [1,2,3]
var.append(4)


#정렬
var = [1,4,3,2]
var.sort() # 오름차순
print(var)
var.reverse() # 내림차순
print(var)

# 인덱스 
var.index(3)

# 추가 (위치,value)
var = [1,2,3]
var.insert(0,4)
print(var)

var = list(range(10,60+1,10))
var.remove(40)
print(var)

# 요서 삭제후 끄집어내기
var = [1,2,3]
print(var.pop())
print(var)
```

```python
# for문을 사용한 pop
var = [1,2,3]

backVar = []
for _ in range(3):
  backVar.append(var.pop())
print(var)
print(backVar)


var = [1,2,3]
var.pop(1)
var.pop(1)





```

```python
# 리스트의 요소 갯수 세기
var = [1,2,3,4,4,4,5]
# 4가 몇개?
print(var.count(4))
# 전체 데이터 갯수
print(len(var))



# 리스트의 확장 (리스트를 합침)
var = [1,2,3,4]
var.extend([5,6])
print(var)

# 1부터 100까지의 리스트 생성
var = list(range(1,100+1))
print(f"Length: {len(var)}")

# var에 101부터 200까지 list를 추가하여 확장
var.extend(list(range(101,200+1)))
print(f"Length: {len(var)}")



```

```python
# 1부터 10까지의 수중 짝수 리스트 만들기
evenVar = list(range(2,10+1,2))
print(evenVar)
# 홀수 리스트 만들고 짝수리스트에 확장
evenVar.extend(list(range(1,9+1,2)))
print(evenVar)
#정렬
evenVar.sort()
print(evenVar)

# 합계
sum(evenVar)
```

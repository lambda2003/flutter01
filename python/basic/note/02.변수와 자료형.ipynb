### 변수
계산에 필요한 자료를 일시적으로 저장하거나 처리결과를  담을 수 있는 기억장소를 지정해 주는 역할
변수 이름은 값을 지정하는 메모리 역역의 이름으로 할당.

### 변수 이름의 작성 규칙.
- 첫자는 영문으로 시작.
- 첫자를 제외하고 숫자, under bar, dot을 사용할수 있음.
- 대문자와 소문자를 구분한다
- 변수 이름은 의미를 파악할수 있는 이름으로 지정하는 것이 좋다.
- 두 단어를 포함한 Camel로 하거나 Snake로 하는것이 가독성이 좋다.

```python
# 연산자
var01 = 3
var02 = 4
print('덧셈:',var01+var02)
print('뺄셈:',var01-var02)
print("실수 나눗셈:",var01/var02)
print("정수 나눗셈 몫", var01//var02)
print("정수 나머지:",var01%var02)
print('제곱:',var01**2)
```

```python
# 거듭제곰
print("거듭제곱: ", var01**2)
print("거듭제곱: ", pow(var01,2))
```

```python
# 파이썬 모듈으르 이용한 수학 계산
import math

print('거듭제곱: ',math.pow(var01,2))
print('제곱근: ',math.sqrt(var01))
print('싸인: ', math.sin(45))

# sin(45) ** 2 + cos(45) **2
print("싸인:",math.sin(45),"코사인:",math.cos(45),"탄젠트: ", math.tan(45),sep=' ')
math.pow(math.sin(45),2) + math.pow(math.cos(45),2)
```

### 관계연산자

```python
num1 = 100
num2 = 20

# 동등비교
print(num1 == num2)
print(num1 != num2)

# 크기 비교
print(num1 <= num2)
print(num1 >= num2)
```

### 논리 연산자

```python
print((num1 >= 50) and (num2 <= 10))
print((num1 >= 50) or (num2 <= 10))

# 부정 연산자
not(num1 >= 50)

```

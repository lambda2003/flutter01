### match case

```python
x = int(input('숫자를 입력하세요:'))
match x%2 :
  case 0:
    print(f"{x}는 짝수 입니다.")
  case 1:
    print(f"{x}는 홀수 입니다.")
```

```python
x = int(input('숫자를 입력하세요:'))

match x:
  case x if x%2 == 0:
    print(f"{x}는 짝수 입니다.")
  case _:
    print(f"{x}는 홀수 입니다.")
```

#### while else
- else문은 반복문이 정상적으로 종료 되었을 경우 실행된다. 
- while문 안에 break가 있어 반복이 중간에 멈추었을 경우 else문은 실행되는 않는다.

```python
i = 1
while i < 5:
  print(i)
  i+=1
else:
  print('반복을 처음부터 끝까지 하였습니다.')
```

```python
i = 1
while i < 5:
  print(i)
  if i == 2: break
  i+=1
else:
  print('반복을 처음부터 끝까지 하였습니다.')
```

```python
# 문장과 단어 추출

string = """나는 홍길동 입니다.
주소는 서울시 입니다.
나이는 35세 입니다.
"""

sents = []
words = []
for sen in string.split("\n"):
  sents.append(sen)

# sents[:len(sents)-1]
del sents[-1]


for sen in sents:
  for word in sen.split(" "):
    words.append(word)

print("문장: ",sents)
print('문장수:',len(sents))
print('단어: ',words)
print('단어수:',len(words))
```

### for else
- for 반복문이 모두 실행되고 나면 else문이 실행된다.
- for 중간에 break로 종료되면 else는 실행되지 않는다. 


```python

for i in range(1,5+1):
  if i== 10:
    break
  print(i)
else:
  print('찾고자 하는 10이 없습니다')



words
```

```python
fruits = ['apple','banana','cherry','orange']

for i,fruit in enumerate(fruits):
  if fruit == 'orange':
    print(f'orange를 찾았습니다.{i}')
    break
else:
  print('orange가 없습니다.')
```

### Python의 Comprehension(내포)

```python
var01 = [1,2,3,4]
var02 = [d*5 for d in var01]
print(var02) #[5, 10, 15, 20]

# var01에서 짝수만 리스트에 넣어 출력
var03 = [d*5 for d in var01 if d%2==0]
print(var03)

```

```python
# 구구단 (2단 ~ 9단) 결과만 리스트로 저장하기
result = [tuple(i*k for k in range(1,9+1)) for i in range(2,9+1)]
print(result)
```

```python
# 60점 이상의 학생 점수만 List로 만들어 평균 구하기

scores = [90,25,67,45,80]
passList = [score for score in scores if score>=60]
print(f"합격한 학생들의 평균 점수는 {np.mean(passList)} 입니다.")
```

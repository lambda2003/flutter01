### 집합(Set) 자료형
- 집합자료형은 집합에 관련된 것들을 쉽게 처리.
  데이터베이스가 나온이후 잘 사용안함.
- 중복을 허용하지 않는다. 
- 순서가 없다.

```python
# Set의 형성

s0 = {1,2,3,4}
s0
```

```python
s1 = set([1,2,3,4,5,6])
s2 = set([4,5,6,7,8,9])

# 교집합
print(s1 & s2)

# 합집합
print(s1 | s2)

# 차집합
print(s1 - s2)
```

---
### 집합관련 함수

```python
# 요소 1개 추가
s1 = set([1,2,3])
s1.add(4)
print(s1)

# 요소 여러개 추가
s1.update([4,5,6])
print(s1)

# 특정값제거
s1.remove(6)
print(s1)

```

```python
import math
# 집합의 사용
# 중복제거 
var = [1,1,2,2,3,3,4,4]
s1 = set(var)
print(list(s1))

sum(list(set(tuple(var))))
math.sqrt(sum(list(set(tuple(var)))))
```

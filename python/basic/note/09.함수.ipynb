### numpy 내장함수
- 빠르다
- 수치해석(numerical method) 이 다들어있다. math보다 큼.

```python
import numpy as np

```

```python
# 통계에서 
# 평균: 제일 많이 있는 곳
print(np.mean(range(1,10+1)))

# 합계
print(np.sum(range(1,10+1)))

# 표준편차 구하기 - 평균선과의 거리에 절대값을 구한것
print(np.std(range(1,10+1)))

# 분산 구하기 - 평균선과의 거리에 제곱을 한값 : 각 분포의 분산값이 같을수가 없다. 
print(np.var(range(1,10+1)))

# 분산에 루트를 적용한 값이 표준편차
np.sqrt(np.var(range(1,10+1))) == np.std(range(1,10+1))

# 결측치 (missing value) : 비어있는 부분
# 이상치 (outlier) : 오류로 잘못 입력한 부분


# 최대값
print(np.max(range(1,10+1)))

# 최소값
print(np.min(range(1,10+1)))

# 중앙값 5,6의 중앙값 
print(np.median(range(1,10+1)))

# 누적합
print(np.cumsum(range(1,10+1)))
print(np.cumsum(range(1,10+1))[-1])
# 누적곱
print(np.cumprod(range(1,10+1)))






```

```python
# 매개 변수가 있는 함수

def calc(num1, num2):
  return (num1+num2, num1-num2,num1*num2,num1/num2)

# 덧셈,뺄셈,곱셈,나눗셈
a,b,c,d = calc(1,2)
print(a,b,c,d)

result = calc(5,6)
print(result)

print(result[0],result[1],result[2],result[3])





```

```python
# 가변 길이 인수를 가진 함수 (Unpacking 연산자)

def sumMany(*args):
  sum = 0
  try: 
    for i in args:
      sum += i
  except Exception as e:   
    sum=False
    raise Exception('ERROR')
  return sum

print(sumMany(1,2))
print(sumMany(1,2,3))
print(sumMany(1,2,3,4))


```

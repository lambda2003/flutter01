### 딕셔너리 (Dictionary) = Dart의 map
: Dictionary는 List나 Tuple처럼 순차적으로 해당 요소값을 구하지 않고 Key를 통해 value를 얻는다.

```python
# 딕셔너리 생성하기
dic = {
  'name':'petty',
  'phone': '010-111-2222',
  'birth' : '1231'
}

print(dic['name'])

# Dictionary의 value가 list인경우
dic = {
  'a':[1,2,3]
}

print(dic['a'])

# 요소 추가하기
dic['b'] = [4,5,6]

```

```python
dic = {
  'name':'petty',
  'phone': '010-111-2222',
  'birth' : '1231'
}

# 키값들
print(dic.keys())
print(list(dic.keys()))

# value값들
print(list(dic.values()))

# items
print(dic.items())

for k,v in dic.items():
  print(f"{k} : {v}")


print('name' in dic)
print('mail' in dic)

# clear dictionary
dic.clear()



```

```python
# Ex: 나의 이름은 홍길동이고 나이는 30세
d = {
  'name':'홍길동',
  'age': 30
}

print(f"나의 이름은 '{d['name']}'이고 나이는 '{d['age']}'세 입니다.")



```

```python
list01 = [1,2,3]
list02 = [1,2,3]

# is 는 python에만 있음
# 파이썬의 메모리 구조를 이해야함.


# 두 객체가 동일한 메모리에 위치?
print(list01 is list02)
# 두 객체의 value가 같냐?
print(list01 == list02)

# 외 true? data저장소는 (data type:값) 이런식으로 저장됨. 
# list01[0] 은 value가 아니라 data type 이 같은지를 물어보는 것임 
print(list01[0] is list02[0])


var01 = [1,2,3]
var02 = var01
print(var01 is var02)

var01 = [1,2,3]
var02 = var01.copy()
print(var01 is var02)

```

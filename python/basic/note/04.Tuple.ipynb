### 튜플(Tuple) 자료형
: 튜플은 다음과 같은 점을 제외하곤 리스트와 거의 비슷하다.
- ( 와 ) 로 사용
- 리스트는 값을 생성, 삭제, 수정이 가능하지만 튜플은 값을 변경할수 없다.
프로그램 작성시 생선된 데이터값을 변화 시키면서 사용하고자 하면 리스트를 사용.
한번 생성된 데이터값이 변경되는 것을 막고자 하면 튜플을 사용.
프로그램 작성시 평균적으로 튜플보다 리스트를 많이 사용.

```python
# Tuple의 선언
var01 = ()
var02 = (1,)
var03 = 1,
var04 = (1,2,3)
var05 = 1,2,3
var06 = ('a','b',('ab','cd'))

print(type(var01))
print(type(var02))
print(var03)

print(var06[-1][0])
```

```python
# Tuple값 변경
## Tuple에 LIST를 추가하여 사용
tuple01 = ([1,2],3,4)
tuple01[0][0] = 100
# tuple01[1] = 200
print(tuple01)
```
